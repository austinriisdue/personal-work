```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
```

## Ex. 6.1

```python
rho = 0.9
T = 10**4
X = [np.random.normal()]
for t in range(1, T):
    et = np.random.normal()
    X.append(rho*X[t-1] + et)
plt.hist(X, bins=80,density=True)
fit = np.random.normal(0, 1 / np.sqrt((1 - rho**2)), size=10**4)
sns.kdeplot(fit)
plt.show()
```
