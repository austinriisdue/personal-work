```python
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
%%time
N = 1000
chips = 100000
local_thresh = 3
grid = {}
all_points_hit = 0
for x in range(-N, N+1):
    for y in range(-N, N+1):
        grid[(x,y)] = 0
grid[(0,0)] = chips
while max(grid.values()) > 3:
    for i in range(-N+1, N-1):
        for j in range(-N+1, N-1):
            while grid[(i,j)] > local_thresh:
                grid[(i - 1, j)] += 1
                grid[(i + 1, j)] += 1
                grid[(i, j - 1)] += 1
                grid[(i, j + 1)] += 1
                grid[(i,j)] -= 4
    all_points_hit += 1
    if np.random.uniform() > 0.99:
        print(max(grid.values()),all_points_hit,(sum(list(1 for x in grid.values() if x == 3)) / len(grid) + sum(list(1 for x in grid.values() if x == 2)) / len(grid) +sum(list(1 for x in grid.values() if x == 1)) / len(grid) + sum(list(1 for x in grid.values() if x == 0)) / len(grid)))
```

```python
plotgrid = {}
for k in grid.keys():
    if abs(k[0]) < 150:
        if abs(k[1]) < 150:
            plotgrid[k] = grid[k]
```

```python
plt.figure(figsize = (15,15))
X = [x[0] for x in plotgrid.keys()]
Y = [x[1] for x in plotgrid.keys()]
color = [grid[(x,y)] for x,y in zip(X,Y)]
plt.scatter(X,Y,c=color,cmap="viridis")
plt.axis('off')
plt.show()
```
