```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
np.random.seed(42069)
```

```python
def mandelbrot(n_rows, n_columns, iterations):
    x_cor = np.linspace(-2,1,n_rows)
    y_cor = np.linspace(-2,1,n_columns)
    x_len = len(x_cor)
    y_len = len(y_cor)
    output = np.zeros((x_len,y_len))
    for i in range(x_len):
        for j in range(y_len):
            c = complex(x_cor[i],y_cor[j])
            z = complex(0, 0)
            count = 0
            for k in range(iterations):
                z = (z * z) + c
                count = count + 1
                if (abs(z) > 4):
                    break
            output[i,j] = count
        #print((i/x_len)*100,"% completed")
    plt.imshow(output.T, cmap = "hsv")
    plt.axis("off")
    plt.show()
mandelbrot(1000,1000,150)
```

```python
def randjulia(n_rows, n_columns, iterations):
    u,v = (0.13651949688697163, -0.5953442489313658)#(st.norm.rvs(0,1/3),st.norm.rvs(0,1/3))
    u,v = (abs(u),-(abs(v)))
    x_cor = np.linspace(-2,1,n_rows)
    y_cor = np.linspace(-2,1,n_columns)
    x_len = len(x_cor)
    y_len = len(y_cor)
    output = np.zeros((x_len,y_len))
    for i in range(x_len):
        for j in range(y_len):
            z = complex(x_cor[i],y_cor[j])
            c = complex(u, v) #-.42,.6 look neat
            count = 0
            for k in range(iterations):
                z = (z * z) + c
                count = count + 1
                if (abs(z) > 4):
                    break
            output[i,j] = count
        #print((i/x_len)*100,"% completed")
    plt.imshow(output.T, cmap = "Greys_r")
    plt.axis("off")
    plt.show()
    return (u,v)
```
