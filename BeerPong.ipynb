```python
import scipy.stats as st
def cupSetup(backrow=4,thirdrow=3,secondrow=2,frontcup=1):
    print(backrow*'  -------   ')
    print(backrow*'  \     /   ')
    print(backrow*'   \___/    ')
    print('       '+thirdrow*'  -------   ')
    print('       '+thirdrow*'  \     /   ')
    print('       '+thirdrow*'   \___/    ')
    print(2*'      '+secondrow*'  -------   ')
    print(2*'      '+secondrow*'  \     /   ')
    print(2*'      '+secondrow*'   \___/    ')
    print(3*'      '+frontcup*'  -------   ')
    print(3*'      '+frontcup*'  \     /   ')
    print(3*'      '+frontcup*'   \___/    ')
```

```python
def shoot(power,backrow,thirdrow,secondrow,frontcup):
    make = 0
    if power == 'low':
        make = st.bernoulli.rvs(.7)
        if make == 1:
            row = st.bernoulli.rvs(.5)
            if row == 0 and frontcup == 1:
                frontcup -=1
                return [backrow,thirdrow,secondrow,frontcup]
            if row == 0 and frontcup == 0:
                row = 1
            if row == 1 and secondrow !=0:
                secondrow -=1
                return [backrow,thirdrow,secondrow,frontcup]
            if row == 1 and secondrow ==0:
                make = 0
        if make == 0:
            return [backrow,thirdrow,secondrow,frontcup]
    if power == 'medium':
        make = st.bernoulli.rvs(.8)
        if make == 1:
            row = st.bernoulli.rvs(.5)
            if row == 0 and secondrow != 0:
                secondrow -=1
                return [backrow,thirdrow,secondrow,frontcup]
            if row == 0 and secondrow == 0:
                row = 1
            if row == 1 and thirdrow !=0:
                thirdrow -=1
                return [backrow,thirdrow,secondrow,frontcup]
            if row == 1 and thirdrow ==0:
                make = 0
        if make == 0:
            return [backrow,thirdrow,secondrow,frontcup]
    if power == 'high':
        make = st.bernoulli.rvs(.8)
        if make == 1:
            row = st.bernoulli.rvs(.5)
            if row == 0 and thirdrow != 0:
                thirdrow -=1
                return [backrow,thirdrow,secondrow,frontcup]
            if row == 0 and thirdrow == 0:
                row = 1
            if row == 1 and backrow !=0:
                backrow -=1
                return [backrow,thirdrow,secondrow,frontcup]
            if row == 1 and backrow ==0:
                make = 0
        if make ==0:
            return [backrow,thirdrow,secondrow,frontcup]
```

```python
def BeerPong():
    backrow = 4
    thirdrow = 3
    secondrow = 2
    frontcup = 1
    cupSetup()
    while backrow+thirdrow+secondrow+frontcup != 0:
        power = input("Select low, medium or high power to shoot with: ")
        while power not in ['low','medium','high']:
            print("Cannot interpret your input.")
            power = input("Select low, medium or high power to shoot with: ")
        newCups = shoot(power,backrow,thirdrow,secondrow,frontcup)
        backrow,thirdrow,secondrow,frontcup = newCups[0],newCups[1],newCups[2],newCups[3]
        totalCups = backrow+thirdrow+secondrow+frontcup
        if totalCups == 6:
            backrow,thirdrow,secondrow,frontcup = 0,3,2,1
            cupSetup(backrow,thirdrow,secondrow,frontcup)
        elif totalCups == 3:
            backrow,thirdrow,secondrow,frontcup = 0,0,2,1
            cupSetup(backrow,thirdrow,secondrow,frontcup)
        else:
            cupSetup(backrow,thirdrow,secondrow,frontcup)
    return "Winner"
```

```python
def BeerPongTurn(backrow,thirdrow,secondrow,frontcup):
    startCups = backrow+thirdrow+secondrow+frontcup
    cupSetup(backrow,thirdrow,secondrow,frontcup)
    power = input("Select low, medium or high power to shoot with: ")
    while power not in ['low','medium','high']:
        print("Cannot interpret your input.")
        power = input("Select low, medium or high power to shoot with: ")
    newCups = shoot(power,backrow,thirdrow,secondrow,frontcup)
    backrow,thirdrow,secondrow,frontcup = newCups[0],newCups[1],newCups[2],newCups[3]
    totalCups = backrow+thirdrow+secondrow+frontcup
    if startCups == totalCups:
        made = 0
    else:
        made = 1
    if totalCups == 6:
        backrow,thirdrow,secondrow,frontcup = 0,3,2,1
        cupSetup(backrow,thirdrow,secondrow,frontcup)
    elif totalCups == 3:
        backrow,thirdrow,secondrow,frontcup = 0,0,2,1
        cupSetup(backrow,thirdrow,secondrow,frontcup)
    else:
        cupSetup(backrow,thirdrow,secondrow,frontcup)
    return [backrow,thirdrow,secondrow,frontcup,made]
```
