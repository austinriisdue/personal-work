```python
import pandas as pd
import scipy.stats as st
from sklearn import neighbors
from scipy.spatial import distance
import numpy as np
```

```python
adpi = pd.DataFrame()
adpi["Name"] = ["Laurie","Sarah","Marcela","Sophia"]
n = len(adpi)
adpi["A"] = [st.bernoulli.rvs(0.5) for x in range(n)]
adpi["B"] = [st.bernoulli.rvs(0.5) for x in range(n)]
adpi["C"] = [st.bernoulli.rvs(0.5) for x in range(n)]
adpi["D"] = [st.bernoulli.rvs(0.5) for x in range(n)]
adpi
```

```python
knn = neighbors.KNeighborsClassifier(n_neighbors = n, metric = 'jaccard')
knn.fit(adpi[[x for x in adpi.columns if x != "Name"]],adpi["Name"])
```

```python
test = [[st.bernoulli.rvs(.5) for x in range(len(adpi.columns) - 1)] for x in range(n-1)]
test
```

```python
err = 0
sis = []
for i in range(len(test)):
    Dj = knn.kneighbors([test[i]])[0][0]
    bijection_index = knn.kneighbors([test[i]])[1][0]
    j = 0
    while j != len(sis):
        if adpi.Name[bijection_index[j]] in sis:
            j+=1
    sis.append(adpi.Name[bijection_index[j]])
    err += Dj[j]
```

```python
adpi_names = adpi_dummies["Name"]
rushee_names = rush_dummies["Name"]
Dj = pd.DataFrame()
Dj["Rushee Name"] = rushee_names
for x in adpi_names:
    dis = []
    for y in rushee_names:
        dis.append(distance.jaccard(adpi_dummies.loc[adpi_dummies["Name"] == x],rush_dummies.loc[rush_dummies["Name"] == y]))
    Dj[x] = dis
```
